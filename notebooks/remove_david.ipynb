```python
from pathlib import Path
import json


def get_all_text(data):
    text = ""
    if isinstance(data, dict):
        for key, value in data.items():
            if key == "Text":
                text += value + "  "
            elif isinstance(value, (dict, list)):
                text += get_all_text(value)
    elif isinstance(data, list):
        for item in data:
            text += get_all_text(item)
    return text


venues = Path("../data/processed/adobe_extracted").iterdir()
common_urls = [
    "yahoo",
    "p.m",
    "a.m",
    "p.a",
    "i.e",
    "e.g",
    "a.b.c",
    "gmail",
    "youtube",
    "vimeo",
    "google",
    "goo.gl",
    "bit.ly",
    "mediazilla",
]

for venue in list(venues):
    try:
        data = json.load(open(venue / "structuredData.json"))

    except:
        continue
    text = get_all_text(data)
    if "maxwell" in text.lower():
        print(venue)
```
