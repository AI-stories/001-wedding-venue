```python
import re
from pathlib import Path

import numpy as np

pdf_dir = Path("../data/raw/Southern California Wedding Venues")
pdf_files = [f.name.replace(".pdf", "") for f in pdf_dir.glob("*.pdf")]

clean_words = [
    "&",
    "the",
    "-",
    "ratesheet",
    "_",
    r"\(",
    r"\)",
    r"\.",
    "merged",
    "pricing",
    r"\d+",
    r"\.pdf",
    "packages",
    "brochure",
    "rate",
    "sheet",
    "information",
    "venue",
    "prices",
    "weddings",
    "wedding",
    "resort",
    "inclusive",
    "package",
    "digital",
    "guide",
    "park",
    "course",
    "golf",
    "banquets",
    "banquet",
    "restaurant",
    "catering",
    "event",
    "planner",
    "garden",
    "hotel",
    "spa",
    "at",
    "santa barbara",
    "santa monica",
    "santa maria",
    "san",
    "marino",
    "diego",
    "los",
    "angeles",
    "and",
    "or",
    "the",
    "of",
    "in",
    "at",
    "to",
    "from",
    "juan",
]


def _clean(name):
    for word in clean_words:
        name = re.sub(word, " ", name.lower())
    name = re.sub(" +", " ", name)
    return name


def _count_overlapping_words(name, venues):
    clean_venues = [_clean(venue) for venue in venues]
    clean_name = _clean(name)
    return [
        sum([word in clean_venue for word in clean_name.split(" ")])
        / len(clean_name.split(" "))
        for clean_venue in clean_venues
    ]


def _find_best_venue(name, venues):
    scores = _count_overlapping_words(name, venues)
    idx = np.argmax(scores)
    return venues[idx], scores[idx]


venues = venue_info["Venue name"].str.lower().tolist()
venues = [_clean(v) for v in venues]
clean_pdf_files = [_clean(f) for f in pdf_files]

df = pd.DataFrame()

df["pdf_name"] = clean_pdf_files
df["og_pdf_name"] = pdf_files
df["venue"], df["ratio"] = zip(*[_find_best_venue(f, venues) for f in pdf_files])
df
```

```python
import re

import numpy as np
from fuzzywuzzy import fuzz
from tqdm import tqdm


def clean_pdf_name(pdf_name):
    remove_words = [
        r"[\-_.&\d]",
        "events",
        ".pdf",
        "weddings",
        "wedding",
        "venue",
        "packages",
        "digital",
        "packet",
        "package",
        "ratesheet",
        "2024",
        "2025",
        "2023",
        "brochure",
    ]

    for word in remove_words:
        pdf_name = re.sub(word, "", pdf_name.lower())
    return pdf_name


venues = np.array([clean_pdf_name(v) for v in venue_info["Venue name"].tolist()])
df2 = pd.DataFrame()
for pdf in tqdm(pdf_files):
    clean_name = clean_pdf_name(pdf)
    ratios = [fuzz.partial_ratio(clean_name, venue) for venue in venues]
    max_ratio = np.max(ratios)
    max_idx = np.argmax(ratios)
    df2 = pd.concat(
        [
            df2,
            pd.DataFrame(
                {
                    "pdf_name": pdf,
                    "venue": venues[max_idx],
                    "url": venue_info["Website "].iloc[int(max_idx)],
                    "ratio": max_ratio,
                },
                index=[0],
            ),
        ]
    )
df2.sort_values(by="ratio", ascending=False)
```
